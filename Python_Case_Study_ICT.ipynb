**Bank Acount Manager**


*   Simulate a bank account with deposit, withdrawal, and balance
enquiry operations
*   Use variables and data types to store account information


*   Implement conditional statements to check for sufficient balance
before withdrawal
*   Use functions to perform deposit, withdrawal, and balance inquiry
operations.

```python
# Account information
account_holder_name = "Anzila"
account_number = 985632
balance = 5000

print("Account Holder Name:", account_holder_name)
print("Account Number:", account_number)
print("Current Balance:", balance)
```

```python
# Deposit Function
def deposit(amount):
  global balance
  balance += amount
  print("Deposited:", amount)
  print("Updated Balance:", balance)

deposit(5000)
```

```python
# Withdraw Function
def withdraw(amount):
    global balance

    if amount <= balance:
        balance -= amount
        print("Withdrawn:", amount)
        print("Remaining Balance:", balance)
    else:
        print("Insufficient Balance!")

withdraw(4000)
```

**Weather Data Analyzer**


*   Analyze weather data (temperature, humidity, etc.) for a given region
*   Use variables and data types to store weather data

*  Implement conditional statements to determine weather conditions
(sunny, rainy, etc.)
*   Use functions to calculate average temperature and humidity, and print
results

```python
temperatures = [30, 32, 29, 31, 33]
humidity = [60, 65, 70, 55, 68]

print("Temperature Data:", temperatures)
print("Humidity Data:", humidity)
```

```python
# Average Temperature
def average_temperature(temp_list):
    avg_temp = sum(temp_list) / len(temp_list)
    return avg_temp

avg_temp = average_temperature(temperatures)
print("Average Temperature:", avg_temp)
```

```python
# Average Humidity
def average_humidity(humidity_list):
    avg_humidity = sum(humidity_list) / len(humidity_list)
    return avg_humidity

avg_humidity = average_humidity(humidity)
print("Average Humidity:", avg_humidity)
```

```python
def weather_condition(temp, hum):
    if temp > 32 and hum < 60:
        print("Weather Condition: Sunny")
    elif hum > 65:
        print("Weather Condition: Rainy")
    else:
        print("Weather Condition: Cloudy")

weather_condition(avg_temp, avg_humidity)
```
